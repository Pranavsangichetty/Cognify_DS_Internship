**Explore the distribution of categorical variables like "Country Code", "City" and "Cuisines"**

```python
cat_cols = ['Country Code', 'City', 'Cuisines']

for col in cat_cols:
    if col in resturant_df.columns:
        print(f"\n {col} ")
        print(f"Unique values: {resturant_df[col].nunique()}")
        print(f"Most common categories:\n{resturant_df[col].value_counts().head(10)}\n")
    else:
        print(f"\n Column '{col}' not found in dataset.")
```

**Identify the top cuisines and cities with the highest number of restaurants**

```python
# For cuisines

from collections import Counter

all_cuisines = resturant_df['Cuisines'].dropna().str.split(',').sum()
all_cuisines = [c.strip() for c in all_cuisines]
all_cuisines
```

```python
cuisine_counts = Counter(all_cuisines)
cuisine_counts
```

```python
cuisine_df = pd.DataFrame(cuisine_counts.items(), columns=['Cuisine', 'Count']).sort_values(by='Count', ascending=False)
cuisine_df
```

```python
top_10_cuisines = cuisine_df.head(10)
top_10_cuisines
```

```python
# For cities

from collections import Counter

all_cities = resturant_df['City'].dropna().str.split(',').sum()
all_cities = [c.strip() for c in all_cities]
all_cities
```

```python
cities_counts = Counter(all_cities)
cities_counts
```

```python
cities_df = pd.DataFrame(cities_counts.items(), columns=['City', 'Count']).sort_values(by='Count', ascending=False)
cities_df
```

```python
top_10_cities = cities_df.head(10)
top_10_cities
```
