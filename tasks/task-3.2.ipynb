## Task-2 : Customer Preference Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data

resturant_df = pd.read_csv(r"C:\Users\sangi\Downloads\Dataset .csv")
resturant_df
```

**Analyze the relationship between the type of cuisine and the restaurant's rating**

```python
resturant_df['Aggregate rating'].dropna()
```

```python
resturant_df['Cuisines'].dropna()
```

```python
# Spliting the resturants with multiple cuisines

resturant_df['Cuisines'] = resturant_df['Cuisines'].apply(lambda x: x.split(',')[0].strip() if isinstance(x, str) else x)
resturant_df['Cuisines']
```

```python
from collections import Counter

cuisine_rating = resturant_df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()

# Count number of restaurants per cuisine

cuisine_count = resturant_df['Cuisines'].value_counts().reset_index()
cuisine_count.columns = ['Cuisines', 'Restaurant Count']
cuisine_count
```

```python
# Merging cuisine's count and rating

cuisine_merge = pd.merge(cuisine_rating, cuisine_count, on='Cuisines')

# Sort by average rating

cuisine_merge = cuisine_merge.sort_values(by='Aggregate rating', ascending=False)
cuisine_merge.head(10)
```

```python
# Identifying popular and high rated cuisines

popular_cuisines = cuisine_merge[cuisine_merge['Restaurant Count'] > 10]

# Finding top cuisines by rating

top_cuisines = popular_cuisines.sort_values(by='Aggregate rating', ascending=False).head(10)
top_cuisines
```

**Identify the most popular cuisines among customers based on the number of votes**

```python
resturant_df['Votes'].dropna()
```

```python
# Grouping cuisine and calculating by total votes

cuisine_votes = resturant_df.groupby('Cuisines')['Votes'].sum().reset_index()
    
# Sort in descending order of votes
cuisine_votes = cuisine_votes.sort_values(by='Votes', ascending=False)
cuisine_votes.head(10)
cuisine_votes
```

**Determine if there are any specific cuisines that tend to receive higher ratings**

```python
# Calculating mean and count of ratings for each cuisine

cuisine_rating_summary = resturant_df.groupby('Cuisines').agg(
    Average_Rating=('Aggregate rating', 'mean'),
    Restaurant_Count=('Cuisines', 'count')
).reset_index()

cuisine_rating_summary = cuisine_rating_summary[cuisine_rating_summary['Restaurant_Count'] > 10]

# Sorted by average rating

cuisine_rating_summary = cuisine_rating_summary.sort_values(by='Average_Rating', ascending=False)
cuisine_rating_summary.head(10)
```
