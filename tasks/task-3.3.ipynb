## Task-3 : Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data

resturant_df = pd.read_csv(r"C:\Users\sangi\Downloads\Dataset .csv")
resturant_df
```

**Create visualizations to represent the distribution of ratings using different charts (histogram, bar plot, etc.)**

```python
resturant_df['Aggregate rating'].dropna()
```

```python
# Histogram

plt.figure(figsize=(8,6))
sns.histplot(resturant_df['Aggregate rating'], bins=10, kde=True, color='skyblue', edgecolor='black')
plt.title("Distribution of Restaurant Ratings", fontsize=14, fontweight='bold')
plt.xlabel("Aggregate Rating", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Bar Plot

plt.figure(figsize=(8,6))
sns.countplot(x='Aggregate rating', data=resturant_df, hue='Aggregate rating', 
              palette='viridis', legend=False)
plt.title('Distribution of Restaurant Ratings', fontsize=14)
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Box Plot

plt.figure(figsize=(8,6))
sns.boxplot(x='Price range', y='Aggregate rating',
            data=resturant_df,
            hue='Price range',           
            palette='coolwarm',
            legend=False)                 
plt.title('Aggregate Rating by Price Range', fontsize=14)
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=resturant_df, alpha=0.6, color='teal', edgecolor='black')
plt.title('Relationship Between Votes and Ratings', fontsize=14)
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
# Pie Chart

plt.figure(figsize=(7, 7))
resturant_df['Price range'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    cmap='Set3',
    textprops={'fontsize': 12}
)
plt.title('Distribution of Restaurants by Price Range', fontsize=14)
plt.ylabel('')
plt.show()
```

**Compare the average ratings of different cuisines or cities using appropriate visualizations**

```python
# Considering average rating of top 10 cuisines 

top_cuisines = resturant_df['Cuisines'].value_counts().head(10).index
avg_rating_cuisine = resturant_df[resturant_df['Cuisines'].isin(top_cuisines)] \
    .groupby('Cuisines')['Aggregate rating'].mean().reset_index() \
    .sort_values(by='Aggregate rating', ascending=False)

# Plotting

plt.figure(figsize=(10,6))
sns.barplot(data=avg_rating_cuisine, 
            x='Cuisines', 
            y='Aggregate rating', 
            hue='Cuisines', 
            palette='viridis', 
            legend=False)

plt.title('Average Restaurant Ratings by Top 10 Cuisines', fontsize=14)
plt.xlabel('Cuisine Type')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Considering average rating of top 10 cities

top_cities = resturant_df['City'].value_counts().head(10).index
avg_rating_city = resturant_df[resturant_df['City'].isin(top_cities)] \
    .groupby('City')['Aggregate rating'].mean().reset_index() \
    .sort_values(by='Aggregate rating', ascending=False)

# Plotting

plt.figure(figsize=(10,6))
sns.barplot(data=avg_rating_city, 
            x='City', 
            y='Aggregate rating', 
            hue='City', 
            palette='coolwarm', 
            legend=False)

plt.title('Average Restaurant Ratings by Top 10 Cities', fontsize=14)
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Visualize the relationship between various features and the target variable to gain insights**

```python
# Relationship between Votes and Rating

plt.figure(figsize=(8,6))
sns.scatterplot(data=resturant_df, 
                x='Votes', 
                y='Aggregate rating', 
                hue='Price range', 
                palette='viridis', 
                alpha=0.7)
plt.title('Votes vs Aggregate Rating (by Price Range)', fontsize=14)
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.legend(title='Price Range')
plt.show()
```

```python
# Relationship between Price Range and Rating

plt.figure(figsize=(8,6))
sns.boxplot(data=resturant_df, 
            x='Price range', 
            y='Aggregate rating', 
            hue='Price range', 
            palette='coolwarm', 
            legend=False)
plt.title('Aggregate Rating by Price Range', fontsize=14)
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
# Relationship between City and Rating

top_cities = resturant_df['City'].value_counts().head(10).index
avg_rating_city = resturant_df[resturant_df['City'].isin(top_cities)] \
    .groupby('City')['Aggregate rating'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=avg_rating_city, 
            x='City', 
            y='Aggregate rating', 
            hue='City', 
            palette='coolwarm', 
            legend=False)
plt.title('Average Aggregate Rating by Top 10 Cities', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Relationship between Cuisine and Rating

top_cuisines = resturant_df['Cuisines'].value_counts().head(10).index
avg_rating_cuisine = resturant_df[resturant_df['Cuisines'].isin(top_cuisines)] \
    .groupby('Cuisines')['Aggregate rating'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=avg_rating_cuisine, 
            x='Cuisines', 
            y='Aggregate rating', 
            hue='Cuisines', 
            palette='magma', 
            legend=False)
plt.title('Average Rating by Top 10 Cuisines', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Relationship between all numeric features

numeric_df = resturant_df.select_dtypes(include=['number'])
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Between Numerical Features', fontsize=14)
plt.show()
```
